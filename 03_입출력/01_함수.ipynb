```python
# 함수란 무엇인가? 

# def 함수이름(매개변수1, 매개변수2 ...):
#     수행할문장1
#     수행할문장2
#     ...

def add(a, b):
    return a + b 

```

```python
print(add(3, 4))
```

```python
# 1. 일반적인 함수 
# 입력값 O, 반환값 O

def add(a, b):
    result = a + b    
    return result 

c = add(3, 4)
print(c)
```

```python
# 2. 입력값이 없는 함수 
# 입력값 X, 반환값 O
def say():
    return "hi"


c = say()
print(c)
```

```python
# 3. 반환값이 없는 함수
# 입력값 O, 반환값 X

def add(a, b): 
    print(a + b)

c = add(3, 4)
print(c)
```

```python
# 4. 입력값도 없고, 반환값도 없는 함수
# 입력값 X, 반환값 X

def say():
    print("hi")


c = say()
print(c)
```

```python
# 매개변수를 지정하여 호출하기

def sub(a, b): 
    return a - b


c = sub(b=3, a=7)
print(c)
```

```python
# 입력값이 몇 개가 될지 모를때는 어떻게 해야 할까? 

print("hi", "hello", "bye")
```

```python
def add_many(*args):
    result = 0
    print("args : ", args)
    print("args : ", type(args))
    for i in args:
        result = result + i
    return result 


c = add_many(1, 2, 3, 4, 5)
print(c)
```
