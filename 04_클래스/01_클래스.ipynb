```python
class Calculator:
    def __init__(self): 
        self.result = 0 

    def add(self, num):
        self.result = self.result + num
        return self.result 
    
    def sub(self, num): 
        self.result -= num 
        return self.result 
    
    def mul(self, num):
        self.result *= num
        return self.result 
    
    def div(self, num): 
        self.result /= num 
        return self.result 
    

    

cal3 = Calculator()
cal3.add(7)
cal3.sub(4)
cal3.mul(2)
cal3.div(3)
```

```python
# 객체에 연산할 숫자 지정하기 
class FourCal:
    def setdata(self, first, second): 
        self.first = first 
        self.second = second 
    
    def add(self):
        result = self.first + self.second
        return result 
    
    def sub(self): 
        result = self.first - self.second 
        return result 
    
    def mul(self): 
        result = self.first * self.second 
        return result 

    def div(self):
        result = self.first / self.second 
        return result 


cal4 = FourCal()
cal4.setdata(3, 4)
print(cal4.first)
print(cal4.second)

cal5 = FourCal()
cal5.setdata(10, 20)

print(cal5.add())
print(cal5.sub())
print(cal5.mul())
print(cal5.div())
```
