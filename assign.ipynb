````python
### 문제 3: 1부터 N까지의 합 
n = int(input("숫자를 입력하세요: "))

if n > 0:
    i = 1
    total = 0
    expr = ""
    while i <= n:
        total += i # total = total + i 
        expr += str(i) # expr = expr + str(i) -> "" + "1" -> "1 + " + "2"
        if i != n:
            expr += " + " # expr = expr + " + " -> "1 + " -> "1 + 2" + " + "
        i += 1
    print(f"{expr} = {total}")
else:
    print("양수만 입력 가능합니다.")
````

````python
## 문제 4: 홀수/짝수 분류기
odds = []
evens = []

for i in range(1, 21):
    if i % 2 == 0:
        evens.append(i)
    else:
        odds.append(i)

print("홀수: " + ", ".join(map(str, odds)) + f" (총 {len(odds)}개)")
print("짝수: " + ", ".join(map(str, evens)) + f" (총 {len(evens)}개)")
````

````python
### 문제 6: 숫자 피라미드
height = int(input("높이를 입력하세요 (1~10): "))

if 1 <= height <= 10:
    for i in range(1, height + 1):
        # 앞쪽 공백
        print(" " * (height - i), end="")
        # 증가하는 부분
        for j in range(1, i + 1):
            print(j, end="")
        # 감소하는 부분
        for j in range(i - 1, 0, -1): # 
            print(j, end="")
        print()
else:
    print("높이는 1에서 10 사이의 정수만 입력 가능합니다.")
````

````python
num = int(input("숫자를 입력하세요: "))
primes = []

is_prime = True
if num < 2: is_prime = False

for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
        is_prime = False
        break

print(f"{num}은(는) {'소수입니다.' if is_prime else '소수가 아닙니다.'}")

for i in range(2, num + 1):
    check = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            check = False
            break
    if check: primes.append(i)

print(f"\n1부터 {num}까지의 소수:\n{', '.join(map(str, primes))}")
print(f"총 {len(primes)}개의 소수가 있습니다.")
````

````python
num = int(input("2 이상의 숫자를 입력하시오: "))

prime_list = [] # 소수들을 담을 바구니
prime = False # 입력받은 숫자가 소수인지 저장할 변수

if num >= 2:
    # 2부터 입력받은 숫자까지 n을 이용하여 하나씩 소수인지 검사
    for n in range(2, num + 1):
        is_prime = True # 현재 n이 일단 소수라고 가정
        
        # n이 소수인지 확인하기 위해 2부터 (n-1)까지 나눠봄
        for i in range(2, n):
            if n % i == 0:
                is_prime = False # 나누어떨어지면 소수가 아님
                break
        
        # for문 이후 True가 유지되어 소수로 판별되었다면 리스트에 추가
        if is_prime:
            prime_list.append(str(n))
            # 지금 True인 숫자가 입력한 숫자라면? 입력받은 숫자 소수 판별 변수(prime) 결과 True로 변경
            if n == num:
                prime = True

    # 1. 입력받은 숫자의 소수 여부 출력
    if prime:
        print(f"{num}은(는) 소수입니다.")
    else:
        print(f"{num}은(는) 소수가 아닙니다.")

    # 2. 1부터 입력받은 숫자까지의 모든 소수 출력
    print(f"1부터 {num}까지의 소수:")
    print(", ".join(prime_list))
    
    # 3. 소수의 개수 출력
    print(f"총 {len(prime_list)}개의 소수가 있습니다.")

else:
    print("숫자가 2보다 작습니다. 다시 입력하세요.")
````

````python
### 문제 8: 간단한 ATM 시스템
# 간단한 ATM 기능을 구현하세요.

# **요구사항**:
# - while 문으로 메뉴 반복
# - 초기 잔액: 10,000원
# - 기능: 1.잔액조회 2.입금 3.출금 4.종료
# - 출금 시 잔액 부족 체크

# **실행 예시**:
# ```
# === ATM 시스템 ===
# 1. 잔액 조회
# 2. 입금
# 3. 출금
# 4. 종료
# 선택: 1
# 현재 잔액: 10,000원

# 선택: 2
# 입금액: 5000
# 입금 완료! 현재 잔액: 15,000원

# 선택: 3
# 출금액: 20000
# 잔액이 부족합니다. 현재 잔액: 15,000원

# 선택: 4
# 이용해 주셔서 감사합니다.
# ```
money = 10000

while money :
    print("=== ATM 시스템 ===")
    print("1. 잔액조회")
    print("2. 입금")
    print("3. 출금")
    print("4. 종료")
    input_num = int(input("선택:"))
    if input_num == 1:
        print(f"현재잔액 : {money}원")
    elif input_num == 2:
        input_money = int(input("입금액:"))
        money = money + input_money
        print(f"입금완료! 현재잔액 : {money}원")
    elif input_num == 3:
        input_money = int(input("출금액:"))      
        if money < input_money:
            print(f"잔액이 부족합니다. 현재잔액 : {money}원")
        else :
            money = money - input_money
            print(f"출금완료! 현재잔액 : {money}원")
    elif input_num == 4:
        print("이용해주셔서 감사합니다.")
        break

````

````python
### 문제 9: 업다운 게임 (20분)
# 컴퓨터가 1~100 사이의 랜덤 숫자를 생성하고, 사용자가 맞추는 게임을 만드세요.

# **요구사항**:
# - `import random` 사용
# - while 문으로 게임 진행
# - 시도 횟수 카운트
# - Up/Down 힌트 제공

# **실행 예시**:
# ```
# 1부터 100사이의 숫자를 맞춰보세요!

# 시도 1: 50
# Up!

# 시도 2: 75
# Down!

# 시도 3: 60
# Down!

# 시도 4: 55
# 정답입니다! 4번 만에 맞추셨습니다!
# ```
import random

random_num = random.randint(1, 100)
print("1부터 100사이의 숫자를 맞춰보세요!")

count = 1
while random_num:
    input_num = int(input(f"시도 {count}:"))
    if input_num > random_num:
        print("Down!\n")
        count += 1
    elif input_num < random_num:
        print("Up!\n")
        count += 1
    else:
        print(f"정답입니다! {count}번 만에 맞추셨습니다")
        break
````

````python
# ### 문제 12: 간단한 쇼핑몰 장바구니
# 상품을 장바구니에 담고, 총 금액을 계산하는 프로그램을 작성하세요.

# **요구사항**:
# - 최소 5개의 상품과 가격 설정
# - 메뉴: 1.상품목록 2.장바구니담기 3.장바구니보기 4.결제 5.종료
# - 할인 쿠폰 적용 기능 (10% 할인)
# - 30,000원 이상 구매 시 배송비 무료

# ---
products = {
    "사과":5000,
    "바나나":3000,
    "치킨":20000,
    "삼겹살":10000,
    "양파":1000,
    "초코파이":4000,
    "만두":9000,
}
    
cart = []

while True:
    print("\n[메뉴]")
    print("1. 상품목록")
    print("2. 장바구니 담기")
    print("3. 장바구니 보기")
    print("4. 결제")
    print("5. 종료")

    menu = input("선택: ")

    # 1. 상품목록
    if menu == "1":
        print("\n[상품 목록]")
        for name, price in products.items():
            print(f"{name}: {price}원")

    # 2. 장바구니 담기
    elif menu == "2":
        item = input("상품 이름 입력: ")
        if item in products:
            cart.append(item)
            print(f"{item} 담기 완료!")
        else:
            print("없는 상품입니다.")

    # 3. 장바구니 보기
    elif menu == "3":
        if not cart:
            print("장바구니가 비어 있습니다.")
        else:
            print("\n[장바구니]")
            total = 0
            for item in cart:
                price = products[item]
                print(f"{item}: {price}원")
                total += price
            print(f"총 금액: {total}원")

    # 4. 결제
    elif menu == "4":
        if not cart:
            print("장바구니가 비어 있습니다.")
            continue

        total = sum(products[item] for item in cart)

        # 쿠폰 적용 여부
        coupon = input("쿠폰 사용 (y/n): ")
        if coupon == "y":
            total *= 0.9
            print("10% 할인 적용!")

        # 배송비
        if total >= 30000:
            delivery = 0
        else:
            delivery = 3000

        final_total = int(total + delivery)

        print(f"상품 금액: {int(total)}원")
        print(f"배송비: {delivery}원")
        print(f"최종 결제 금액: {final_total}원")

        cart.clear()  # 결제 후 장바구니 비우기

    # 5. 종료
    elif menu == "5":
        print("프로그램 종료")
        break

    else:
        print("잘못된 입력입니다.")
````
